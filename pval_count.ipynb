```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
path = '/home/pitikovegor/CourseWork/genomes/reds_experiments'
cell_type = 'K562'
feat_type = 'gene'
```

```python
table = pd.read_csv(f'{path}/{cell_type}/id_to_gene_{feat_type}.tsv', sep='\t', header = None)
```

```python
table
```

```python
table = table.drop_duplicates()
```

```python
table = table.groupby([4, 6, 5]).size()
table = pd.DataFrame(table)
table.reset_index(inplace=True)
```

```python
normalised = pd.read_csv('/home/pitikovegor/CourseWork/genomes/normalisation_data/K562.pvalue.tab.txt', sep='\t')
```

```python
big_table = pd.merge(left=normalised, right=table, left_on=['gene_name', 'ensg'], right_on=[4, 6], how ='inner')
```

```python
table
```

```python
big_table = big_table[['gene_name', 'ensg', 'gene_chr', 'gene_start', 'gene_end', 'rnaseq_count', 0, 'gene_type']]
```

```python
big_table.loc[:, 'cpr'] = big_table.loc[:, 0]/big_table.loc[:, 'rnaseq_count']
```

```python
big_table
```

```python
all_con = list(big_table['cpr'])
all_con.sort()
lnc = list(big_table[big_table['gene_type']=='lncRNA']['cpr'])
lnc.sort()
prot = list(big_table[big_table['gene_type']=='protein_coding']['cpr'])
prot.sort()
```

```python
len(all_con)
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

sns.distplot(all_con[:-200], hist = True, kde = False, rug = False,
             color = 'green', 
             hist_kws = {'alpha': 1.0},
             kde_kws={'linewidth': 1.5},
             rug_kws={'color': 'red'},
             bins = 500)



sns.distplot(prot[:-100], hist = True, kde = False, rug = False,
             color = 'blue',
             hist_kws = {'alpha': .7},
             kde_kws={'linewidth': 1.5},
             rug_kws={'color': 'blue'},
             bins = 500)

sns.distplot(lnc[:-100], hist = True, kde = False, rug = False,
             color = 'red', 
             hist_kws = {'alpha': .7},
             kde_kws={'linewidth': 1.5},
             rug_kws={'color': 'red'}, 
             bins = 500)
plt.legend(labels=['All', 'protein_coding', 'lncRNA'])
# Plot formatting
plt.title('Distribution of contacts per RNA')
plt.xlabel('Contacts per RNA')
plt.ylabel('Dencity')
plt.xlim(0, 30)
#plt.ylim(0, 1)
plt.savefig(f'/home/pitikovegor/CourseWork/genomes/pictures/{cell_type}/contacts_count_hist_all.png', dpi=600, format='png')
```
